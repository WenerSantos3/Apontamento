```python
# Instale o moviepy, caso não tenha
!pip install moviepy

import requests
from moviepy.editor import VideoFileClip, concatenate_videoclips
import os

# Links dos vídeos
videos = {
    'vinheta': 'https://Bambino.b-cdn.net/vignettes/default_vignette.mp4',
    'opening': 'https://Bambino.b-cdn.net/openings/default_opening.mp4',
    'introduction': 'https://Bambino.b-cdn.net/introductions/default_introduction.mp4',
    'content': 'https://bambino.b-cdn.net/Series/16008120-a282-4457-bc44-3dceac1ab7f4/tmp_3n10jhy.mp4',
    'credit': 'https://Bambino.b-cdn.net/credits/default_credits.mp4'
}

# Função para baixar vídeo
def download_video(url, filename):
    response = requests.get(url, stream=True)
    with open(filename, 'wb') as f:
        for chunk in response.iter_content(chunk_size=8192):
            if chunk:
                f.write(chunk)

# Baixar todos os vídeos
for name, url in videos.items():
    print(f"Baixando {name}...")
    download_video(url, f"{name}.mp4")

# Carregar vídeos com moviepy
clips = []
for name in ['vinheta', 'opening', 'introduction', 'credit']:
    print(f"Carregando {name}.mp4...")
    clip = VideoFileClip(f"{name}.mp4")
    clips.append(clip)

# Concatenar vídeos
final_clip = concatenate_videoclips(clips, method="compose")

# Salvar vídeo final
print("Salvando final_video.mp4 ...")
final_clip.write_videofile("final_video.mp4", codec="h264_nvenc", audio_codec="aac")

# Liberar recursos e remover arquivos temporários
for clip in clips:
    clip.close()
for name in videos:
    os.remove(f"{name}.mp4")

print("Processo concluído! Vídeo final salvo como final_video.mp4")
```

### Run ComfyUI with cloudflared (Recommended Way)




```python
!wget https://github.com/cloudflare/cloudflared/releases/latest/download/cloudflared-linux-amd64.deb
!dpkg -i cloudflared-linux-amd64.deb

import subprocess
import threading
import time
import socket
import urllib.request

def iframe_thread(port):
  while True:
      time.sleep(0.5)
      sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
      result = sock.connect_ex(('127.0.0.1', port))
      if result == 0:
        break
      sock.close()
  print("\nComfyUI finished loading, trying to launch cloudflared (if it gets stuck here cloudflared is having issues)\n")

  p = subprocess.Popen(["cloudflared", "tunnel", "--url", "http://127.0.0.1:{}".format(port)], stdout=subprocess.PIPE, stderr=subprocess.PIPE)
  for line in p.stderr:
    l = line.decode()
    if "trycloudflare.com " in l:
      print("This is the URL to access ComfyUI:", l[l.find("http"):], end='')
    #print(l, end='')


threading.Thread(target=iframe_thread, daemon=True, args=(8188,)).start()

!python main.py --dont-print-server
```

### Run ComfyUI with colab iframe (use only in case the previous way with localtunnel doesn't work)

You should see the ui appear in an iframe. If you get a 403 error, it's your firefox settings or an extension that's messing things up.

If you want to open it in another window use the link.

Note that some UI features like live image previews won't work because the colab iframe blocks websockets.

```python
import threading
def iframe_thread(port):
  while True:
      time.sleep(0.5)
      sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
      result = sock.connect_ex(('127.0.0.1', port))
      if result == 0:
        break
      sock.close()
  from google.colab import output
  output.serve_kernel_port_as_iframe(port, height=1024)
  print("to open it in a window you can open this link here:")
  output.serve_kernel_port_as_window(port)

threading.Thread(target=iframe_thread, daemon=True, args=(8188,)).start()

!python main.py --dont-print-server
```
